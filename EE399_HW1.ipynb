```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt

X=np.arange(0,31)
Y=np.array([30, 35, 33, 32, 34, 37, 39, 38, 36, 36, 37, 39, 42, 45, 45, 41, 40, 39, 42, 44, 47, 49, 50, 49, 46, 48, 50, 53, 55, 54, 53])

e2 = None

def velfit(c, x, y):
    e2 = np.sqrt(np.sum((c[0]*np.cos(c[1]*x)+c[2]*x+c[3]-y)**2))
    return e2

v0 = np.array([3, 1*np.pi/4, 2/3, 32])

res= opt.minimize(velfit, v0, args=(X, Y), method='Nelder-Mead')

c = res.x


print(f'Param: A={c[0]}, B={c[1]}, C={c[2]}, D={c[3]}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.arange(0, 31)
Y = np.array([30, 35, 33, 32, 34, 37, 39, 38, 36, 36, 37, 39, 42, 45, 45, 41,
              40, 39, 42, 44, 47, 49, 50, 49, 46, 48, 50, 53, 55, 54, 53])

B_values = np.linspace(0.01, 0.5, 50) #sweep these paramaters
C_values = np.linspace(-2, 2, 50)

A = 25 # keep these two parameters the same
D = 50

def f(x, A, B, C, D):
    error_grid = np.zeros((len(B_values), len(C_values)))
    for i, B in enumerate(B_values):
        for j, C in enumerate(C_values):
            error_grid[i, j] = np.sqrt(np.mean((A*np.cos(B*x) + C*x + D - Y)**2))
    return error_grid

B_grid, C_grid = np.meshgrid(B_values, C_values) # creating a grid to print the errors on
plt.pcolor(B_grid, C_grid, error_grid, cmap='jet')
plt.xlabel('B')
plt.ylabel('C')
plt.title('Error landscape')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

X = np.arange(0, 21)
Y = np.array([30, 35, 33, 32, 34, 37, 39, 38, 36, 36, 37, 39, 42, 45, 45, 41,
              40, 39, 42, 44, 47])

coefficients_line = np.polyfit(X, Y, deg=1)
line = np.polyval(coefficients_line, X)
error_line = np.sum((line - Y)**2)
print(f'error for line: {error_line}')

coefficients_parabola = np.polyfit(X, Y, deg=2)
parabola = np.polyval(coefficients_parabola, X)
error_parabola = np.sum((parabola - Y)**2)
print(f'error for parabola: {error_parabola}')

coefficients_poly19 = np.polyfit(X, Y, deg=19)
poly19 = np.polyval(coefficients_poly19, X)
error_poly19 = np.sum((poly19 - Y)**2)
print(f'error for 19th degree polynomial: {error_poly19}')

plt.scatter(X, Y, label='Data')
plt.plot(X, line, label='Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Line fit')
plt.show()

plt.scatter(X, Y, label='Data')
plt.plot(X, parabola, label='Parabola')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Parabola fit')
plt.show()

plt.scatter(X, Y, label='Data')
plt.plot(X, poly19, label='Poly19')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Poly19 fit')
plt.show()
```
